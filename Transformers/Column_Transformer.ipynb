### Column Transformer
* Transformation of columns before passing to our model
* Simple Imputer - for missing values columns
* Ordinal Encoding - for columns which has some order
* One Hot Encoding - for columns which does not have any order
* Label Encoding - for output column .
* Remainder - for the left our columns

```python
import pandas as pd
import numpy as np
import sklearn
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('E:/csvfiles/covid_toy.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
transformer = ColumnTransformer(transformers=[
    ('transformer1',SimpleImputer(),['fever']),
    ('transformer2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('transformer3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
```

```python
transformer.transform(x_test)
```
