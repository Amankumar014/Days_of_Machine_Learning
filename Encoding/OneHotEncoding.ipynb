#### OneHotEncoding is used when our data is Nominial and not Ordinal .
* Here we make columns for each category in our data
* OHE by default uses sparse Matrix
#### Dummy Variable Trap
* After OHE , we get as many no of columns as there are categories in our data say 'n' . After this we drop any one columns and therefore our dataset now has 'n-1' columns for our OHE columns .
* We do this because of Multicollinearity i.e there should not be any relationship among our OHE columns as it may create problems in our ML problems such as Linear REgression , Logistic Regression etc . So to erode the relationship between our columns if present any we remove column .
* So in conclusion we represent 'n' categories with 'n-1' columns in OHE.
